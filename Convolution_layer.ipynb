```python
import torch
import  torch.nn as nn
```

```python
# Khởi tạo một DL ảnh xám có kích thước 6*6 với chanel = 1
image = torch.randint(high=5, size=(1,6,6),dtype=torch.float32)
image
```

```python
# Khởi tạo convolution layer với 3 kernel kích thước 3*3 và giá trị của kernel được khởi tạo ngẫu nhiên 
conv_layer = nn.Conv2d(
    in_channels=1, # Số chanel của ảnh đầu vào
    out_channels=3, # số lượng future map đầu ra - số chanel đầu ra
    kernel_size=3, # kích thước của kernel
    stride=1, # bước nhảy của kernel
    bias=True,
    padding=1
)
conv_layer.weight.shape
```

```python
conv_layer.weight
```

```python
future_maps = conv_layer(image)
future_maps
```

```python
future_maps.shape
```

```python
Pooling_Layer = nn.MaxPool2d(
    kernel_size=2,
    stride=2)
```

```python
Pooling_Layer(future_maps)
```

```python
Avg_Pooling = nn.AvgPool2d(
    kernel_size=2,
    stride=2
)
```

```python
Avg_Pooling(future_maps)
```
